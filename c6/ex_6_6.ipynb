```r
data <- read.csv("ex_6_6.csv", header=TRUE, row.names=1); data
```

```r
d <- as.dist(data); d    # dist: fviz_dend(res.hc, rect = TRUE)距离
```

## (1) 最短距离

```r
h1 <- hclust(d, "single")
```

```r
plot(h1, hang=-1, main="Shortest Distance Cluster", xlab="Language")    # hang=-1 表示线画到底
```

## (2) 最长距离

```r
h2 <- hclust(d, "complete")
plot(h2, hang=-1, main="Longest Distance Cluster", xlab="Language")
```

## (3) 类平均距离

```r
h3 <- hclust(d, "average")
plot(h3, hang=-1, main="Classes Average Distance Cluster", xlab="Language")
```

## (4) 重心距离

```r
h4 <- hclust(d, "centroid")
plot(h4, hang=-1, main="Centroid Distance Cluster", xlab="Language")
```
