```r
data <- read.csv("ex_6_7.csv", row.names=1); data
```

```r
d <- dist(data)
```

## (1) 最长距离法

```r
h1 <- hclust(d, method="complete")

plot(h1, hang=-1, main="Longest Distance Cluster", xlab="Year")    # hang=-1 表示线画到底
rect.hclust(h1, k=3)    # 画出分为 3 类的矩形框
```

```r
# HELPER：Years in the hclust order
cat(h1$labels[h1$order], sep="，")
```

可以看出聚成三类的结果是：

- 第一类：1994，1999，2000，1995，1998，1996，1997
- 第二类：1988，1987，1985，1986
- 第三类：1993，1989，1990，1991，1992

## (2) 类平均距离法

```r
h2 <- hclust(d, method="average")

plot(h2, hang=-1, main="Classes Average Distance Cluster", xlab="Year")
rect.hclust(h2, k=3)    # 画出分为 3 类的矩形框
```

```r
# HELPER：Years in the hclust order
cat(h2$labels[h2$order], sep="，")
```

聚成三类的结果是：

- 第一类：1994，1999，2000，1995，1998，1996，1997
- 第二类：1987，1985，1986
- 第三类：1993，1988，1992，1991，1989，1990

## (3) 数据标准化

### 数据标准化

先用 scale 将数据标准化：

```r
data.scale <- scale(data); data.scale
```

重新计算标准化数据的距离：

```r
ds <- dist(data.scale)
```

### 最长距离法

```r
h3 <- hclust(ds, method="complete")

plot(h3, hang=-1, main="Longest Distance Cluster - scaled data", xlab="Year")
rect.hclust(h3, k=3)
```

```r
# HELPER：Years in the hclust order
cat(h3$labels[h3$order], sep="，")
```

聚成三类的结果：

- 第一类：1994，1999，2000，1995，1998，1996，1997
- 第二类：1988，1987，1985，1986
- 第三类：1993，1992，1989，1990，1991

### 类平均距离法

```r
h4 <- hclust(ds, method="average")

plot(h4, hang=-1, main="Classes Average Distance Cluster - scaled data", xlab="Year")
rect.hclust(h4, k=3)
```

```r
# HELPER：Years in the hclust order
cat(h4$labels[h4$order], sep="，")
```

聚成三类的结果：

- 第一类：1994，1999，2000，1995，1998，1996，1997
- 第二类：1988，1987，1985，1986
- 第三类：1993，1992，1989，1990，1991

可以看到，在数据标准化之前不同聚类方法得到的结果不仅尽相同，而且在标准化前后聚类结果也是不要一样的，但是在数据标准化之后，两种不同的聚类方法得到了相同的结果。
